```python
mpPose = mp.solutions.pose
pose = mpPose.Pose()
mpDraw = mp.solutions.drawing_utils # For drawing keypoints
points = mpPose.PoseLandmark # Landmarks
```

```python
def labeller(path,f_num):
  cap = cv2.VideoCapture(path)
  total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
  print(total_frames)
  frame=[]
  for fno in range(0, total_frames,1):
    #print(fno)
    temp=[]
    cap.set(cv2.CAP_PROP_POS_FRAMES, fno)
    _, image = cap.read()
    img = image.copy()
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = pose.process(imgRGB)
    landmark=results.pose_landmarks.landmark
    r=[]
    for i in landmark:
      r.append([i.x,i.y])
    r=np.array(r)
    r=r.flatten()
    y=0
    if(fno>=f_num):
      y=1
    else:
      y=0
    l=np.append(r,y)
    l=l.flatten()
    frame.append(l)
  return frame
```

```python
frame = labeller('ggg/ayan_crossing_210_189.mp4', 210)
```

```python
def setdata(frame):
  f=pd.DataFrame(frame)
  f.columns=[f"x{i}" for i in range(1,68)]
  f1 = f.rename(columns={'x67':'y'})
  return f1
```
