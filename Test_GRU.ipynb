```python
import keras
```

```python
model = keras.models.load_model('saved_model_latest/GRU')

# Check its architecture
model.summary()
```

```python
import cv2
import mediapipe as mp
import numpy as np
```

```python
mpPose = mp.solutions.pose
pose = mpPose.Pose()
mpDraw = mp.solutions.drawing_utils # For drawing keypoints
points = mpPose.PoseLandmark # Landmarks
```

```python
def getarray(landmark): 
  r=[]
  for i in landmark:
    r.append([i.x,i.y])
  r=np.array(r)
  r=r.flatten()
  return r
```

```python
def test(path, name):
  cap = cv2.VideoCapture(path)
  width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH) + 0.5)
  height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT) + 0.5)
  size = (width, height)
  fourcc = cv2.VideoWriter_fourcc(*'XVID')
  out = cv2.VideoWriter(f'{name}.avi', fourcc, 30.0, size)
  total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
  print(total_frames)
  arr=[]
  for fno in range(0, total_frames,1):
    print(f"frame number is{fno}")
    #print(fno)
    cap.set(cv2.CAP_PROP_POS_FRAMES, fno)
    _, image = cap.read()
    img = image.copy()
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = pose.process(imgRGB)
    landmark=results.pose_landmarks.landmark
    r=getarray(landmark)
    #print(r)
    if(fno<14):
      arr.append(r)
    elif(fno==14):
      arr.append(r)
      k=np.array(arr).reshape(-1,15,66)
      print(model.predict(k)[0])
    else:
      del arr[0]
      arr.append(r)
      k=np.array(arr).reshape(-1,15,66)
      p=model.predict(k)[0][0]
#       if((fno+1)%30==0):
      p=round(p,2)
      x = "Crossing" if p>0.5 else "Not Crossing"
      color = (0, 0, 255) if p > 0.5 else (0, 255, 0)
      indent = (400,325) if p < 0.5 else (450,325)
      p=str(p)
      print(p)
      cv2.putText(image, p, (500,300), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), thickness = 2)
      cv2.putText(image, x, indent, cv2.FONT_HERSHEY_TRIPLEX, 1, color, thickness = 2)
      cv2.imshow('test',image)
      if cv2.waitKey(10) & 0xFF==ord('q'):
          break
      out.write(image)

    #out.write(image)

  cap.release()
  cv2.destroyAllWindows()
```

```python
test("../test/preetam saha not crossing the road.mp4", "Preetam_final_GRU")
```
